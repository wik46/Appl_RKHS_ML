## Hyperplanes in R3
Example of a program that inputs a vector and scalar and plots a hyperplane

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python

def plot_hyperplane(normal_vector, scalar):
    # Define a grid of points
    xx, yy = np.meshgrid(range(-10, 11), range(-10, 11))
   
    # Calculate the z-coordinate for each point on the grid using the hyperplane equation
    zz = (-normal_vector[0] * xx - normal_vector[1] * yy - scalar) * 1.0 / normal_vector[2]

    # Create a 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Plot the hyperplane
    ax.plot_surface(xx, yy, zz, alpha=0.5)

    # Plot the normal vector
    ax.quiver(0, 0, 0, normal_vector[0], normal_vector[1], normal_vector[2], color='red')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Hyperplane with Normal Vector ' + str(normal_vector) + ' and Scalar ' + str(scalar))

    plt.show()
```

```python

# Example usage
normal_vector = np.array([1, 2, -1])  # Define the normal vector of the hyperplane
scalar = 5  # Define the scalar of the hyperplane equation
plot_hyperplane(normal_vector, scalar)
```
